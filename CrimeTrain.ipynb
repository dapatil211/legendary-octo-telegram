```python
import pandas
import numpy

crime=pandas.read_csv("train.csv")

print(crime.describe())
print(list(crime))
print(crime['Dates'].describe())
```

```python
# time of day
def getTime(date):
    time = date.split(' ')[1].split(':')
    return (24*int(time[0]) + int(time[1]))

month_days = {'01':0, '02':31, '03':59, '04':90, '05':120, '06':151, '07':181, '08':212, '09':243, '10':273, '11':304, '12':334}

# gets date including year
def getFullDate(date):
    date = date.split(' ')[0].split('-')
    return (int(date[0])-2003)*365 + month_days[date[1]] + int(date[2])

# gets date disregarding year
def getDate(date):
    date = date.split(' ')[0].split('-')
    return month_days[date[1]] + int(date[2])

crime["Time"] = crime['Dates'].apply(getTime)
crime["FullDate"] = crime['Dates'].apply(getFullDate)
crime['Date'] = crime['Dates'].apply(getDate)
crime
```

```python
# Converting district names to numbers
distNames = crime['PdDistrict'].unique()
districts = crime['PdDistrict']
for dist in range(len(distNames)):
    districts[districts==distNames[dist]] = dist
crime['PdDistrict'] = districts

# Converting crime categories to numbers
catNames = crime['Category'].unique()
categories = crime['Category']
for cat in range(len(catNames)):
    categories[categories==catNames[cat]] = cat
crime['Category'] = categories

# Converting day of week to numbers
days = crime['DayOfWeek']
dayNames = days.unique()
for day in range(len(dayNames)):
    days[days == dayNames[day]] = day
crime
```
